# 1️⃣ Hashing Numbers

```python
def mod(number, cellnumber):
    return number % cellnumber
```

% means remainder.

This always gives a result between 0 and cellNumber - 1.

If you have cellNumber buckets, it picks which one to use.

Example:

mod(400, 24)

→ 400 % 24 = 16

✅ Store it in bucket 16.

# 2️⃣ Hashing Strings

```python
def modASCII(string, cellnumber):
    total = 0
    for i in string:
        total += ord(i)
    return total % cellnumber  
print(modASCII("ABC", 67))  
```

ord(i) gives you the ASCII code for a character.

Add up all ASCII codes → you get a number.

Then take % cellNumber → you get a bucket index.

Example:

For "ABC":

ord('A') = 65, ord('B') = 66, ord('C') = 67

Total = 65 + 66 + 67 = 198

198 % 24 = 6

✅ Store in bucket 6.

# ✅ 1. Find the First Non-Repeating Character

```python
from collections import Counter

class Solution:
    def firstUniqChar(self, s: str) -> int:
        count = Counter(s)
        for i, char in enumerate(s):
            if count[char] == 1:
                return i
        return -1
```

# 217. Contains Duplicate

```python
from typing import List
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()
        for n in nums:
            if n in seen:
                return True
            seen.add(n)
        return False
```

# 1. Two Sum

```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashmap = {}
        for i, num in enumerate(nums):
            comp = target - num
            if comp in hashmap:
                return [hashmap[comp], i]
            hashmap[num] = i
```

# 49. Group Anagrams

```python
from collections import defaultdict
from typing import List

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        hashmap = defaultdict(list)
        for word in strs:
            sorted_word = ''.join(sorted(word))
            hashmap[sorted_word].append(word)
        return list(hashmap.values())
```
