## 1. üî∑ Bubble Sort
## üìå What is Bubble Sort?
Bubble sort is the simplest sorting algorithm, also known as sinking sort.

üîÅ It works by repeatedly swapping adjacent elements if they are in the wrong order, until the list is sorted.

##  üß† How it Works (Step-by-step):
Compare two adjacent elements.

If the left element is greater than the right one, swap them.

Continue for the entire list (this is called a pass).

After each pass, the largest unsorted element "bubbles up" to the correct position.

Repeat the passes until no more swaps are needed.

## üßÆ Time & Space Complexity
Case	Time Complexity	Explanation
Best	O(n)	When the list is already sorted (can be optimized)
Average	O(n¬≤)	Two nested loops
Worst	O(n¬≤)	List sorted in reverse
Space	O(1)	In-place sort (no extra space)

## ‚úÖ When to Use Bubble Sort
When the list is small.

When the list is almost sorted.

When simplicity is preferred over performance.

When space is limited (O(1)).

## ‚ùå When to Avoid
When performance is critical.

For large datasets.

### üî∑ Bubble Sort ‚Äì Step-by-Step Explanation

We are given a list:

```
[5, 9, 3, 1, 2]
```

Our goal is to **sort this list in ascending order** (from smallest to largest) using **Bubble Sort**.

---

### üí° Idea of Bubble Sort

Bubble Sort compares each **pair of adjacent elements** and **swaps them if they are in the wrong order**. This way, the **largest number "bubbles up" to the end** in each pass.

---

### üîÅ First Pass:

1. Compare 5 and 9 ‚Üí No swap  
   ‚ûú `[5, 9, 3, 1, 2]`  
2. Compare 9 and 3 ‚Üí Swap  
   ‚ûú `[5, 3, 9, 1, 2]`  
3. Compare 9 and 1 ‚Üí Swap  
   ‚ûú `[5, 3, 1, 9, 2]`  
4. Compare 9 and 2 ‚Üí Swap  
   ‚ûú `[5, 3, 1, 2, 9]`

‚úÖ 9 is now in the correct position at the end.

---

### üîÅ Second Pass:

1. Compare 5 and 3 ‚Üí Swap  
   ‚ûú `[3, 5, 1, 2, 9]`  
2. Compare 5 and 1 ‚Üí Swap  
   ‚ûú `[3, 1, 5, 2, 9]`  
3. Compare 5 and 2 ‚Üí Swap  
   ‚ûú `[3, 1, 2, 5, 9]`

‚úÖ 5 is now in the correct position before 9.

---

### üîÅ Third Pass:

1. Compare 3 and 1 ‚Üí Swap  
   ‚ûú `[1, 3, 2, 5, 9]`  
2. Compare 3 and 2 ‚Üí Swap  
   ‚ûú `[1, 2, 3, 5, 9]`

‚úÖ 3 is now in the correct position.

---

### üîÅ Fourth Pass:

1. Compare 1 and 2 ‚Üí No swap  
   ‚ûú `[1, 2, 3, 5, 9]`

‚úÖ Already sorted!

---

### ‚úÖ Final Sorted List:

```
[1, 2, 3, 5, 9]
```

---

### üß† Summary

- Bubble Sort compares adjacent elements and swaps them if they are in the wrong order.
- After each pass, the largest remaining element moves to its correct position.
- It continues until the entire list is sorted.



````python
# üî∑ Bubble Sort Implementation in Python

def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n - 1):  # Total n-1 passes
        for j in range(n - i - 1):  # Compare up to unsorted part
            if arr[j] > arr[j + 1]:
                # Swap elements if they are in the wrong order
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Example list
data = [5, 9, 3, 1, 2]

print("Original List:", data)

# Calling the bubble sort function
bubble_sort(data)

print("Sorted List:  ", data)
````
