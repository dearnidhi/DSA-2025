# Analyze a string, count vowels and consonants separately, and print the result neatly in JSON format


```python
import json
def count_vowels_cosonants(s):
  vowels = "aioueAOIUE"
  vowels_count = 0
  consonants_count =0
  for ch in s:
    if ch.isalpha():
      if ch in vowels:
        vowels_count += 1
      else:
        consonants_count += 1
  result ={"vowels_count":vowels_count, "consonants_count":consonants_count}
  return json.dumps(result, indent=4)
print(count_vowels_cosonants("Hello"))
```

#  Longest Substring Without Repeating Characters

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        seen = set()          # Step 1: Create an empty set to store unique characters of the current window
        left = 0              # Step 2: Left pointer of the sliding window (starts from 0)
        max_len = 0           # Step 3: Variable to store the maximum length found so far

        for right in range(len(s)):   # Step 4: Move the right pointer through each character in the string
            while s[right] in seen:   # Step 5: If the current character is already in the set (duplicate found)
                seen.remove(s[left])  #   Remove the character at the left pointer from the set
                left += 1             #   Move the left pointer one step forward
            seen.add(s[right])        # Step 6: Add the current character to the set
            max_len = max(max_len, right - left + 1)  # Step 7: Update the maximum length of substring

        return max_len
```

#  Longest Palindromic Substring

```python
def longestPalindrome(s):
    n = len(s)
    ans = ""

    for i in range(n):
        # Odd length palindrome (like "aba")
        l, r = i, i
        while l >= 0 and r < n and s[l] == s[r]:
            if (r - l + 1) > len(ans):
                ans = s[l:r+1]
            l -= 1
            r += 1

        # Even length palindrome (like "abba")
        l, r = i, i+1
        while l >= 0 and r < n and s[l] == s[r]:
            if (r - l + 1) > len(ans):
                ans = s[l:r+1]
            l -= 1
            r += 1

    return ans
```

# 6. Zigzag Conversion

```python
class Solution(object):
    def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """
        # Edge case: single row means no zigzag
        if numRows == 1 or numRows >= len(s):
            return s

        # Buckets for each row
        rows = [""] * numRows

        curRow = 0
        goingDown = False

        for c in s:
            rows[curRow] += c
            # Top or bottom row â†’ change direction
            if curRow == 0 or curRow == numRows - 1:
                goingDown = not goingDown
            # Move up or down
            curRow += 1 if goingDown else -1

        # Join all rows
        return "".join(rows)
```

# Integer to Roman

```python
class Solution(object):
    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """
        values = [
            1000, 900, 500, 400,
            100, 90, 50, 40,
            10, 9, 5, 4,
            1
        ]
        symbols = [
            "M", "CM", "D", "CD",
            "C", "XC", "L", "XL",
            "X", "IX", "V", "IV",
            "I"
        ]
        roman = []

        for i in range(len(values)):
            while num >= values[i]:
                num -= values[i]
                roman.append(symbols[i])

        return "".join(roman)
```

# 14. Longest Common Prefix


```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs:
            return ""

        # Take the first string as prefix
        prefix = strs[0]

        # Compare with the rest of the strings
        for s in strs[1:]:
            # Keep reducing prefix until it matches the beginning of s
            while not s.startswith(prefix):
                prefix = prefix[:-1]  # shrink from end
                if not prefix:       # if prefix becomes empty, no common prefix
                    return ""

        return prefix
```
